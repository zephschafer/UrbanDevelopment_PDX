### 2 - Read PDF

```python
# https://stackoverflow.com/questions/24844729/download-pdf-using-urllib
import urllib2

def download_file((download_url, i)):
    file_name = "../data/document_" + str(i) + ".pdf"
    response = urllib2.urlopen(download_url)
    file = open(file_name, 'w')
    file.write(response.read())
    file.close()
    print("Completed")
```

```python
download_file(url)
path = "../document_1.pdf"
```

```python
rough_text = re.sub(r'\n','LINE',convert(path))
```

```python
case_num_replace = re.compile('(\w\w)-(\w\w\w\w\w\w)-(\w\w\w)'\
                              '-(\w\w)-(\w\w)')
```

### 3 - Arrange as Cleaned DataFrame

```python
df = pd.read_table('../data/textfile.txt')
```

```python
# Rename raw column
header_info = df.columns.values
header_info[0]
df.rename(columns={header_info[0]: 'raw'}, inplace=True)
```

```python
# Define Case Number Column
raw_str_list = [df['raw'][i] 
                for i in range(0,len(df['raw']))]
case_num_find = re.compile('\w\w-\w\w\w\w\w\w-\w\w\w-\w\w-\w\w')
case_num_dict = {'case_num':[]}
for elem in raw_str_list:
    case_num_dict['case_num'].append(re.findall(case_num_find, elem)[0])
df['case_num'] =pd.DataFrame.from_dict(case_num_dict)
```

```python
# Define Address Column
raw_str_list = [df['raw'][i] 
                for i in range(0,len(df['raw']))]
address_find = re.compile('(LINELINE?)(.*?), \d{5}?')
address_dict = {'address':[]}
for elem in raw_str_list:
    try:
        address_dict['address']\
            .append(re.findall(address_find, elem)[0][1])
    except:
        address_dict['address'].append('ADDRESS_NOT_FOUND')
df['address'] =pd.DataFrame.from_dict(address_dict)
```

```python
try:
    del gdf
except:
    pass
for address in tqdm(df[:]['address']):
    try:
        query = "https://www.portlandmaps.com/arcgis/rest/services/Public/"\
        "COP_OpenData_Property/MapServer/47/query?where="\
        "UPPER(ADD_FULL)=UPPER('" + str(address) + "')"\
        "&text=&objectIds=&time=&geometry=&"\
        "geometryType=esriGeometryEnvelope&inSR=&spatialRel="\
        "esriSpatialRelIntersects&relationParam=&outFields=*&"\
        "returnGeometry=true&returnTrueCurves=false&maxAllowableOffset=&"\
        "geometryPrecision=&outSR=&returnIdsOnly=false&"\
        "returnCountOnly=false&orderByFields=&groupByFieldsForStatistics"\
        "=&outStatistics=&returnZ=false&returnM=false&gdbVersion=&"\
        "returnDistinctValues=false&resultOffset=&resultRecordCount=&"\
        "queryByDistance=&returnExtentsOnly=false&datumTransformation=&"\
        "parameterValues=&rangeValues=&f=geojson"
        b = bytes(requests.get(query).content)
        with fiona.BytesCollection(b) as f:
            crs = f.crs
            temp_gdf = gpd.GeoDataFrame.from_features(f, crs=crs)
            if len(temp_gdf) > 0:
                if ('gdf' in locals()) == False:
                    gdf = temp_gdf
                else:
                    temp_gdf = gpd.GeoDataFrame.from_features(f, crs=crs)
                    if len(temp_gdf) > 0:
                        gdf = gdf.append(temp_gdf)
    except:
        pass
```

```python
# Source <https://gis.stackexchange.com/questions/268250/
# generating-polygon-representing-rough-100km-circle-around-latitude
# -longitude-poi>
counter = 0
for point in gdf['geometry']:
    n_points = 10
    d = 20  # meters
    angles = np.linspace(0, 360, n_points)
    polygon = geog.propagate(point, angles, d)
    if counter == 0:
        gdf['geom_poly'] = shapely.geometry.Polygon(polygon)
    else:
        gdf['geom_poly'][counter] = shapely.geometry.Polygon(polygon)
    counter += 1
gdf2 = gpd.GeoDataFrame(gdf['geom_poly']).set_geometry('geom_poly')
```

```python
gdf2['map_color'] = 'yellow'
gdf2['map_height'] = 200
```

```python
try:
    os.remove('../data/addresses.geojson')
except:
    pass   
gdf2.to_file('../data/addresses.geojson', driver="GeoJSON")
```
